```python
df2 = df.groupby('Continent').mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2.plot()
```

```python
df3 = df2[df.columns[5:13]].transpose()
df3.plot(figsize=(20, 7), marker='o').invert_xaxis() 

plt.title('Total Population Per Year')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.gca()
plt.show()
```
