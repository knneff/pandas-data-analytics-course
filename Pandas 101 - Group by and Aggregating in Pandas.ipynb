# Group by and Aggregating

```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\knmnl\Documents\Git repository\PandasYouTubeSeries-main\Flavors.csv")
```

```python
df
```

```python
(
    df
    .groupby("Base Flavor")
    .agg({
        "Flavor Rating": "mean",
        "Texture Rating": "mean",
        "Total Rating": "mean"
    })
)
```

```python
df.groupby('Base Flavor').max()
```

```python
df.groupby('Base Flavor').min()
```

```python
df.groupby('Base Flavor').sum(numeric_only = True)
```

```python
(
    df
    .groupby('Base Flavor')
    .agg({
        'Flavor Rating': ['sum', 'mean', 'max', 'min'],
        'Texture Rating': ['sum', 'mean', 'max', 'min'],
        'Total Rating': ['sum', 'mean', 'max', 'min']
    }).round(2)
)
```

```python
(
    df
    .groupby('Base Flavor')
    .agg(
        flavor_sum=("Flavor Rating", "sum"),
        flavor_mean=("Flavor Rating", "mean"),
        flavor_max=("Flavor Rating", "max"),
        flavor_min=("Flavor Rating", "min"),

        texture_sum=("Texture Rating", "sum"),
        texture_mean=("Texture Rating", "mean"),
        texture_max=("Texture Rating", "max"),
        texture_min=("Texture Rating", "min"),

        total_sum=("Total Rating", "sum"),
        total_mean=("Total Rating", "mean"),
        total_max=("Total Rating", "max"),
        total_min=("Total Rating", "min"),
    ).round(2)
)
```

```python
(
    df
    .groupby(['Base Flavor', 'Liked'])
    .agg({
        'Total Rating': ['min', 'max', 'mean']
    })
    .round(2)
)
```

```python
df.groupby(['Base Flavor', 'Liked']).describe()
```
